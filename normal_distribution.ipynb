```python
from scipy.stats import norm
import numpy as np

def normal_distribution(mu, sigma, x):
    return norm.pdf(x, mu, sigma)

x=np.linspace(-10,10,100)
y=normal_distribution(0,1,x)
y2=normal_distribution(0,2,x)

import matplotlib.pyplot as plt
plt.plot(x,y, label='mu=0, sigma=1')
plt.plot(x,y2, label='mu=0, sigma=2')
plt.legend()
plt.savefig('normal_distribution.png')
```

sampling 

```python

samples=np.random.normal(0,2,10)
target=normal_distribution(1,2,samples)
```
